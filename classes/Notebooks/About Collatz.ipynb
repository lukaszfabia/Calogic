# Collatz problem

#### The Collatz problem is a famous problem in mathematics. It is also known as the 3n+1 problem. The problem is as follows:
* ##### Take any positive integer n.
* ##### if n is even, divide it by 2 to get n/2. 
* ##### if n is odd, multiply it by 3 and add 1 to get 3n+1.
* ##### repeat the process indefinitely.
* ##### The conjecture is that no matter what number you start with, you will always eventually reach 1.

The name comes from the German mathematician Lothar Collatz who first proposed the problem in 1937. Let's check formula:

$$ f(n) = \begin{cases}
    \frac{n}{2} && \text{if } n \ \text{mod} \ 2 = 0 \\
    3n + 1  && \text{if } n \ \text{mod} \ 2 \neq 0
\end{cases}
 $$
